# UNSUPERVISED LEARNING

```python
# Import libraries
# General 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import seaborn as sns

# For clustering algorithms
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist

# K-means clustering
from sklearn.cluster import KMeans

# Agglomerative hierarchical clustering
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Import data and print the shape of the data
df = pd.read_csv("FitnessTrackerData.txt",delimiter=" ")
print(df.shape)
```

```python
# Explore the type of data and feature names
df.sample(8)
```

```python
# For consistency, make the data X
X = df.copy()
```

## Plot Data


```python
# Plot a three-dimensional plot of the data
fig = plt.figure(figsize=(9,9))
ax = plt.axes(projection="3d")

x_points = X.loc[:,"Distance"]
y_points = X.loc[:,"Frequency"]
z_points = X.loc[:,"Pace"]

ax.set_xlabel("Distance (km)")
ax.set_ylabel("Frequency (runs per 28 days)")
ax.set_zlabel("Pace (min/km)")

ax.scatter3D(x_points, y_points, z_points);

plt.show()
```

```python
# Plot Pace vs Distance
plt.scatter(X.loc[:,"Pace"],X.loc[:,"Distance"])
plt.title('Graph 1: Pace vs Distance')
plt.xlabel('Pace')
plt.ylabel("Distance")
plt.show()
```

```python
# Plot Pace vs Frequency
plt.scatter(X.loc[:,"Pace"],X.loc[:,"Frequency"])
plt.title('Graph 2: Pace vs Frequency')
plt.xlabel('Pace')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot Distance vs Frequency
plt.scatter(X.loc[:,"Distance"],X.loc[:,"Frequency"])
plt.title('Graph 3: Distance vs Frequency')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.show()
```

## K-Means Clustering

### Finding the Optimum Number of Clusters

```python
# 1. Elbow method 
# Calculate distortions 
# Set random_state=0 and use init='k-means++'
distortions = []
for i in range(1, 16):
    km = KMeans(n_clusters=i, init='k-means++', 
                n_init=10, max_iter=300, tol=1e-04, random_state=0)
    km.fit(X)
    distortions.append(sum(np.min(cdist(X, km.cluster_centers_, 
                      'euclidean'),axis=1)) / X.shape[0])
# Plot distortions
plt.plot(range(1, 16), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
# 2. Silhouette method 
sil = []
kmax = 10
### BEGIN SOLUTION
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(X)
  labels = kmeans.labels_
  sil.append(silhouette_score(X, labels, metric = 'euclidean'))
### END SOLUTION

# Plot
plt.plot(range(2, kmax+1), sil, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

###  Cluster the data using the scikit-learn K-means algorithm

```python
# Cluster the data 
# Hint: km = KMeans(....,init='k-means++', random_state=0)
km = KMeans(n_clusters=3, init='k-means++', 
            n_init=10, max_iter=300, 
            tol=1e-04, random_state=0)
km.fit(X)
```

```python
# Add the cluster labels to the dataframe 
# Set random_state=0
Xnew = X.copy()
Xnew["CLUSTERS"] = km.labels_
Xnew.sample(8, random_state=0)
```

###  Plot the clusters generated by the K-means algorithm

```python
# Plot all variables and their clusters using seaborn 
var = ['Pace', 'Distance','Frequency']

# Plot using seaborn
sns.pairplot(Xnew, vars=var, hue="CLUSTERS", palette=sns.color_palette("hls", 3))
```

## Hierarchical clustering

### Find the optimal number of clusters

```python
# Plot the dendrogram with  first choice of linkage method 
# Hint: Use dendrogram()
dendro = dendrogram(linkage(X, method  = "ward"))

# Note that it may take a while for this graph to be generated
plt.title('Dendrogram')
plt.ylabel('Dissimilarity')
plt.xlabel('Observations')
plt.show()
```

```python
# Plot the dendrogram with  second choice of linkage method 
dendro = dendrogram(linkage(X, method  = "complete"))
# Note that it may take a while for this graph to be generated
plt.title('Dendrogram')
plt.ylabel('Dissimilarity')
plt.xlabel('Observations')
plt.show()
```

### Cluster the data using the scikit-learn AgglomerativeClustering algorithm

#### Grouping the Data into 2 clusters

```python
# Run the agglomerative clustering algorithm using two clusters and the linkage method of  choice 
# Hint: n_clusters=2
hc = AgglomerativeClustering(n_clusters = 2, metric = 'euclidean', linkage ='ward')
y_hc=hc.fit_predict(X)
```

```python
# Set up dataframe with cluster labels 
# Set random_state=0
Xnew = X.copy()
Xnew["CLUSTERS"] = y_hc
Xnew.sample(10, random_state=0)
```

#### Plots with 2 clusters

```python
# Plot all the variables and their clusters using seaborn 
var = ["Pace", "Distance","Frequency"]

# Plot
sns.pairplot(Xnew, vars=var, hue="CLUSTERS", palette=sns.color_palette("hls", 2))
```

#### Grouping the Data into 3 clusters

```python
# Run the agglomerative clustering algorithm using three clusters and the linkage method of your choice 
# Hint: n_clusters=3
hc = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage ='ward')

y_hc=hc.fit_predict(X)
```

```python
# Set up dataframe with cluster labels 
# Set random_state=0
Xnew = X.copy()
Xnew["CLUSTERS"] = y_hc
Xnew.sample(10, random_state=0)
```

```python
# Plot all the variables and their clusters using seaborn 
var = ["Pace", "Distance","Frequency"]

# Plot
sns.pairplot(Xnew, vars=var, hue="CLUSTERS", palette=sns.color_palette("hls", 3))
```

#### Insights
#### 2 cluster hierarchical clustering (0 (red), 1 (green))
	The three features pace, distance and frequency are roughly divided into high and low values.
	Cluster 0 has what seems like two sub groups.
#### 3 Cluster (cluster 0(red), cluster 1(green), cluster2(blue))
	All 3 clusters are shown to be compact and well separated across all dimensions.
	The natural groupings in the data are reflected better here.
#### Number of clusters chosen for K-means clustering 
3
#### Best number of clusters
The best number of clusters is 3. Reason for this choice is that with three clusters, visual distinction is evident in every feature combination. In contrast to the two cluster variant, each cluster is rather tight and distinct. In addition to avoiding overgeneralization, the 3 cluster model identifies significant structure in the data.
